```python
import os


def eliminar_archivos_dsc(ruta_carpeta):
    """
    Elimina todos los archivos que terminen en DSC.nii en una carpeta y sus subcarpetas.

    Args:
      ruta_carpeta: Ruta a la carpeta que se desea analizar.
    """

    for directorio, subdirectorios, archivos in os.walk(ruta_carpeta):       
        # Eliminar archivos DSC.nii en el directorio actual y sus subcarpetas.
        # Recorrer el directorio y sus subcarpetas y eliminar los archivos DSC.nii.
        # Si el archivo no es un directorio, eliminarlo. Si es un directorio,
        # recursivamente eliminar los archivos DSC.nii en el directorio.
        # Si el archivo es un directorio, eliminarlo. Si es un archivo,
        # eliminarlo si su nombre termina en DSC.nii.
        for archivo in archivos:
            if archivo.endswith("DSC.nii.gz"):
                ruta_archivo = os.path.join(directorio, archivo)
                os.remove(ruta_archivo)
                print(f"Archivo eliminado: {ruta_archivo}")


ruta_carpeta = "E:\\Download\\UPENN-GBM\\Dataset_30_casos\\train\\new_images\\images_DSC"
print(ruta_carpeta)

eliminar_archivos_dsc(ruta_carpeta)
```
